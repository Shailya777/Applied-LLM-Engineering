# Tokens and Tokenization: The LLM Alphabet

## 1. What is a Token?
Language models do not read English text letter-by-letter, nor do they read word-by-word. They read **Tokens**.

* **Why not letters?** It requires too much context window to process a simple sentence. The model would spend compute power just figuring out what a word is.
* **Why not words?** English has over 170,000 words, plus infinite slang, code, and typos. A dictionary of whole words would be overwhelmingly large and incapable of handling Out-Of-Vocabulary (OOV) words.

Tokens are a middle-ground: common words are single tokens, while complex or rare words are broken down into sub-word chunks.

***

## 2. Byte-Pair Encoding (BPE)
Modern LLMs use an algorithm called **Byte-Pair Encoding (BPE)** to build their vocabulary dictionary.

**How it works:**
1. Start with a vocabulary of individual bytes/characters.
2. Scan the training corpus and find the most frequently adjacent pair of tokens (e.g., "t" and "h").
3. Merge them into a new single token ("th") and add it to the vocabulary.
4. Repeat this process thousands of times until you reach a target vocabulary size (often 50,000 to 128,000 tokens).

***

````python
# Tokenization example using Tiktoken Library:

import tiktoken

def demonstrate_tokenization(text_string, model_name="gpt-4"):
    """
    Demonstrates how a string is converted into integer tokens using OpenAI's BPE tokenizer.
    """
    # Load the encoding for the specified model
    encoding = tiktoken.encoding_for_model(model_name)

    # Encode the text string into token integers
    token_ids = encoding.encode(text_string)

    print(f"Original Text: '{text_string}'")
    print("-" * 50)
    print(f"Total Token Count: {len(token_ids)}\n")

    print("Token ID -> String Representation:")
    for token_id in token_ids:
        # Decode each individual token back to string to show the chunks
        chunk = encoding.decode([token_id])
        print(f"[{token_id:6d}]  ->  '{chunk}'")

# Run the demonstration
sample_text = "Tokenization is incredibly fascinating!"
demonstrate_tokenization(sample_text)
````
