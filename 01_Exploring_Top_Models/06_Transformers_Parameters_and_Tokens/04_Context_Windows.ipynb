# Context Windows and Compute Economics

## 1. The Context Window (Short-Term Memory)
The Context Window is the maximum number of tokens a model can process in a single request. It includes both the prompt (your input) and the completion (the model's output).

If you think of the model's parameters (weights) as its long-term, static memory, the context window is its **short-term, working memory**. Once a conversation exceeds this window, the model physically cannot "remember" the earliest parts of the prompt unless you truncate or summarize them.

***

## 2. The Quadratic Cost of Attention
Why don't all models have infinite context windows? The answer lies in the math of Self-Attention.

In the standard Transformer, every token calculates its relevance to *every other token*.
* If you have 10 tokens, that requires $10 \times 10 = 100$ attention operations.
* If you have 100,000 tokens, that requires $100,000 \times 100,000 = 10,000,000,000$ operations.

This is known as **$O(N^2)$ complexity** for both memory and compute. Doubling the context window size quadruples the computational cost and VRAM (Video RAM) required on the GPUs. (Note: Modern innovations like Ring Attention and FlashAttention have mitigated this, allowing for massive expansions, but the fundamental physics of the bottleneck remain).

***

## 3. Modern API Economics: Input vs. Output Tokens
When you build applications on top of LLM APIs, you are billed by the token. You will notice a consistent pricing model: **Output tokens are always significantly more expensive than Input tokens.**

* **Input Tokens (Prompt):** Fast and cheap to process. The model can process the entire input prompt in parallel across its GPUs in a single forward pass.
* **Output Tokens (Completion):** Slow and expensive. The model operates *auto-regressively*. It must generate token 1, append it to the context, run the entire massive neural network to generate token 2, append it, and so on. This sequential generation ties up highly expensive GPU memory bandwidth.

## 4. Frontier Model Context Window Comparison
Context limits are expanding rapidly. Here is the landscape of frontier models (as of 2026):

| Provider | Model | Max Context Window | Approx. Equivalent In Pages |
| :--- | :--- | :--- | :--- |
| **OpenAI** | GPT-4 Turbo | 128,000 tokens | ~300 pages |
| **Anthropic** | Claude 3 Opus | 200,000 tokens | ~500 pages |
| **Google** | Gemini 1.5 Pro | 2,000,000 tokens | ~5,000 pages |

*Note: Pushing a model to its maximum context limit can sometimes degrade its ability to retrieve information perfectly from the middle of the prompt, a phenomenon known as "Lost in the Middle."*
