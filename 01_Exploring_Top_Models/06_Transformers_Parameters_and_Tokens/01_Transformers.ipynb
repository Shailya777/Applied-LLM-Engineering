# The Transformer Architecture: Foundations of Modern AI

## 1. Introduction and The Problem with RNNs
Before 2017, sequence modeling (like translation) was dominated by Recurrent Neural Networks (RNNs) and Long Short-Term Memory networks (LSTMs). These architectures processed tokens sequentially.

**The Bottleneck:**
* **Sequential Processing:** You could not compute token $t$ until you finished token $t-1$. This prevented parallelization across GPUs.
* **Vanishing Gradients:** As sequences grew longer, the gradients used to update weights during backpropagation would shrink exponentially, effectively giving the model "amnesia" for early parts of a long sequence.

The **Transformer** is a neural network architecture introduced by Google researchers that completely discarded recurrence in favor of an "Attention" mechanism, allowing for full parallelization and theoretical infinite-memory retention across sequences.

***

## 2. Core Mechanism: Self-Attention
The heart of the Transformer is Self-Attention. Instead of processing sequentially, the model looks at *all* tokens simultaneously and learns how much "attention" or importance each token should pay to every other token in the sequence.

To do this, it creates three vectors for each token: **Query (Q)**, **Key (K)**, and **Value (V)**.
* **Query:** What am I looking for?
* **Key:** What do I contain?
* **Value:** What is my actual underlying meaning?

The mathematical operation is scaled dot-product attention:
$$\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V$$
