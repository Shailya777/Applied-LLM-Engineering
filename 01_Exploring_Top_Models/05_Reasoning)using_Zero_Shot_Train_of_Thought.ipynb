## Zero Shot Train of Thought Reasoning using Ollama (llama3.1:8b):

```python
from openai import OpenAI

OLLAMA_BASE_URL = 'http://172.23.37.63:11434/v1'
api_key = 'fake_key'

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key=api_key)

user_prompt = """
Alice puts her top-secret hard drive inside a vault in the server room and locks it. She leaves for lunch.
While she is gone, Bob enters the server room, picks the lock, takes the hard drive, and hides it inside his backpack in the breakroom.
Later, Charlie enters the breakroom to clean, accidentally knocks Bob's backpack over, sees the hard drive fall out, and panics. Charlie hides the hard drive under the breakroom microwave.
Alice returns from lunch and needs her hard drive.

Based on this sequence of events, answer these three questions:
1. Where will Alice look for the hard drive first, and why?
2. When Alice yells that the hard drive is missing, where does Bob *think* it is?
3. Where is the hard drive actually located?

CRITICAL INSTRUCTION: You must think step-by-step. Map out the timeline and the knowledge state of each character before answering the final three questions.
"""

response = ollama.chat.completions.create(model= 'llama3.1:8b',
                                          messages= [{
                                              'role': 'user','content': user_prompt}])

print(response.choices[0].message.content)
```

## Zero Shot Train of Thought Reasoning using OpenAI gpt-4o:
