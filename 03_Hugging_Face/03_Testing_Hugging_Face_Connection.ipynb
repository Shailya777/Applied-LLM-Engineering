## Using Hugging Face:

1. Make a free Hugging Face Account on https://huggingface.co/
2. Go to Profile > Settings > Access Tokens > Write > Create New Token
3. Copy The Token Key.
4. Put Token Key in `.env` file as `HF_TOKEN=token_key`.

```python
import os
from dotenv import load_dotenv
from huggingface_hub import login, whoami

load_dotenv(override= True)

hf_token = os.getenv("HF_TOKEN")

login(token= hf_token, add_to_git_credential= True) # Logging in Hugging Face using Token.

# Checking if the Login is Successful:
try:
    user_info = whoami()
    print(f"✅ Successfully authenticated as: {user_info['name']}")
    print(f"✅ Full Name: {user_info['fullname']}")
    print(f"✅ Email: {user_info['email']}")
except Exception as e:
    print(f'Authentication Failed: {e}')
```
