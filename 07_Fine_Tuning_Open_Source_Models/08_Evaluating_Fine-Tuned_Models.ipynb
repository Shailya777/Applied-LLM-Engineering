# Evaluation of Fine-Tuned Models

## Imports and Constants:

```python
# Imports:
import os
from dotenv import load_dotenv
from tqdm import tqdm
from huggingface_hub import login
import torch
import transformers
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from datasets import load_dataset
from datetime import datetime
from peft import PeftModel
from pricer.util import evaluate
```

```python
# Constants:
BASE_MODEL = 'meta-llama/Llama-3.2-3B'
PROJECT_NAME = 'price'
HF_USER = 'Shailya01'
DATASET_NAME = f'{HF_USER}/items_prompt_lite'


# Using Adapters from RUN-2, at Training Step 500, just before overfitting started:
RUN_NAME ='2026-05-14_09.57.44-lite'
REVISION = 'ad579cb0aa9f5c818508ae284c0b0bc9a9ebd46a'

PROJECT_RUN_NAME = f'{PROJECT_NAME}-{RUN_NAME}'
HUB_MODEL_NAME = f'{HF_USER}/{PROJECT_RUN_NAME}'
```

```python
# Hyper-Parameters for QLORA:
QUANT_4_BIT = True
capability = torch.cuda.get_device_capability()
use_bf16 = capability[0] >= 8
print(use_bf16)
```

## Login to Hugging Face:

```python
load_dotenv(override= True)
hf_token = os.getenv('HF_TOKEN')
login(token= hf_token, add_to_git_credential= True)
```

## Loading Dataset:

```python
dataset = load_dataset(path= DATASET_NAME)
test = dataset['test']
```

```python
print(f'Loaded {len(test)} Test Items')
```

## Loading Tokenizer and Model:

```python
# Quantization Configuration:
if QUANT_4_BIT:
    quant_config= BitsAndBytesConfig(
        load_in_4bit= True,
        bnb_4bit_quant_type= 'nf4',
        bnb_4bit_use_double_quant= True,
        bnb_4bit_compute_dtype= torch.bfloat16 if use_bf16 else torch.float16
    )
else:
    quant_config= BitsAndBytesConfig(
        load_in_8bit= True,
        bnb_4bit_compute_dtype= torch.bfloat16 if use_bf16 else torch.float16
    )
```

```python
# Loading Tokenizer:
tokenizer = AutoTokenizer.from_pretrained(
    pretrained_model_name_or_path= BASE_MODEL,
    trust_remote_code= True
)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side= 'right'

# Model:
base_model = AutoModelForCausalLM.from_pretrained(
    pretrained_model_name_or_path= BASE_MODEL,
    quantization_config= quant_config,
    device_map= 'auto'
)
base_model.generation_config.pad_token_id = tokenizer.pad_token_id
```

## Loading Fine-Tuned Model with Peft:

```python
fine_tuned_model = PeftModel.from_pretrained(
    base_model,
    HUB_MODEL_NAME,
    revision= REVISION
)

print(f'Memory Foot Print of Base Model + Adapters: {fine_tuned_model.get_memory_footprint() / 1e9:.2f}GB')
```

## Architecture of Fine-Tuned Model:

```python
fine_tuned_model
```

## Function to Predict Price of Test Item:

```python
def model_predict(item):
    inputs = tokenizer(item['prompt'], return_tensors='pt').to('cuda')

    with torch.no_grad():
        output_ids = fine_tuned_model.generate(**inputs, max_new_tokens= 8)

    prompt_len = inputs['input_ids'].shape[1]
    generated_ids = output_ids[0, prompt_len:]
    return tokenizer.decode(generated_ids, skip_special_tokens=True)
```

```python
model_predict(test[0])
```

## Fine-Tuned Model Evaluation:

```python
evaluate(function= model_predict,
         data= test,
         size= 200)
```
